```python
import pickle
import pandas as pd
```

```python
car_dict = {
    "Levy": 1399,
    "Manufacturer": "LEXUS",
    "Model": "RX 450",
    "Prod_year": 2010,
    "Category": "Jeep",
    "Leather_interior": "Yes",
    "Fuel_type": "Hybrid",
    "Engine_volume": 3.5,
    "Mileage": 186005,
    "Cylinders": 6.0,
    "Gear_box_type": "Automatic",
    "Drive_wheels": "4x4",
    "Wheel": "Left wheel",
    "Color": "Silver",
    "Airbags": 12
}
```

```python
data = pd.DataFrame([car_dict])
data
```

```python
from datetime import datetime

data['Age'] = datetime.now().year - data['Prod. year']

data = data.drop(columns=['Doors', 'Prod. year'], errors='ignore')
```

```python
data
```

```python
one_hot_columns = ['Leather interior', 'Gear box type', 'Drive wheels', 'Wheel']

with open('../models/one_hot_encoder.pkl', 'rb') as f:
    one_hot_encoder = pickle.load(f)
```

```python
# 4. Label Encode Categorical Columns
label_encode_columns = ['Manufacturer', 'Model', 'Category', 'Fuel type', 'Color']

# Load the saved label encoders
with open('../models/label_encoders.pkl', 'rb') as f:
    label_encoders = pickle.load(f)

# Apply label encoding
for column in label_encode_columns:
    le = label_encoders[column]
    data[column] = le.transform(data[column])
```

```python
data
```

```python
# 5. Scale column
numerical_columns = ['Levy', 'Engine volume', 'Mileage', 'Age']

with open('../models/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
    
data[numerical_columns] = scaler.transform(data[numerical_columns])
```

```python
data.columns
```

```python
with open('../models/model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
import pandas as pd

data_encoded = pd.get_dummies(data)
train_columns = model.feature_names_in_

data_encoded = data_encoded.reindex(columns=train_columns, fill_value=0)

predictions = model.predict(data_encoded)
print(predictions)
```
